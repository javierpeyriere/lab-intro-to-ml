```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check for missing values**

```python
#your code here
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
#your code here

spaceship = spaceship.dropna()
spaceship.head()
```

```python
spaceship.columns
```

```python
spaceship['Transported'] = spaceship['Transported'].apply(lambda x : 0 if x == False else 1)
spaceship
```

**KNN**

K Nearest Neighbors is a distance based algorithm, and requeries all **input data to be numerical.**

Let's only select numerical columns as our features.

```python
#your code here
list_numerical = [ 'Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Transported']
list_categorical = ['PassengerId','HomePlanet', 'CryoSleep', 'Cabin', 'Destination','VIP','Name']

df_num = spaceship[list_numerical]
df_num.rename(columns={'Transported':'target'}, inplace = True)
df_num
```

And also lets define our target.

```python
#your code here
sns.pairplot(df_num, hue="target")
plt.show()
```

**Train Test Split**

Now that we have split the data into **features** and **target** variables and imported the **train_test_split** function, split X and y into X_train, X_test, y_train, and y_test. 80% of the data should be in the training set and 20% in the test set.

```python
#your code here
features = df_num.drop(columns = ["target"])
target = df_num["target"]
```

```python
# from sklearn.model_selection import train_test_split # done above
# from sklearn.neighbors import KNeighborsRegressor


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=10)

# we use this to be able to train the model, we usually set it at 0.8 => 0.2 for the test_size 
# X has the features and Y the target, so we remove
# the target from the test data
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

You need to choose between **Classificator** or **Regressor**. Take into consideration target variable to decide.

Initialize a KNN instance without setting any hyperparameter.

```python
#your code here
# knn = KNeighborsRegressor(n_neighbors=100), was when we were playing with age
knn = KNeighborsClassifier(n_neighbors=100) # we can change the number of neighbours of course
```

Fit the model to your data.

```python
#your code here
knn.fit(X_train, y_train)
```

Evaluate your model.

```python
#your code here
knn.score(X_test, y_test) 
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import precision_score, confusion_matrix, recall_score, f1_score 
#importing the metrics we want
```

```python
pred= knn.predict(X_test)
pred

sns.heatmap(confusion_matrix(y_test, pred), annot = True) # good model finally!!
```
